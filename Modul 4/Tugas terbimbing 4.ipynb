## Soal 1

```python
import matplotlib.pyplot as plt
import numpy as np

#mendifinisikan bobot edge
data = [
    ("Budi", 81.24, 73.66, 95.89, 2023),
    ("Siti", 98.52, 84.86, 88.7, 2024),
    ("Asep", 91.96, 71.03, 79.93, 2022),
    ("Dewi", 87.96, 97.28, 71.91, 2024),
    ("Andi", 74.68, 77.76, 79.33, 2023),
    ("Lestari", 74.68, 89.88, 79.76, 2022),
    ("Joko", 71.74, 79.35, 91.89, 2022),
    ("Rina", 95.99, 85.6, 89.13, 2022),
    ("Fajar", 88.03, 86.4, 96.62, 2024),
    ("Indah", 91.24, 75.55, 84.17, 2023),
    ("Agus", 70.62, 99.09, 73.59, 2022),
    ("Fitri", 99.1, 93.25, 91.4, 2022),
    ("Rudi", 94.97, 98.18, 92.82, 2022),
    ("Maya", 76.37, 96.84, 86.84, 2024),
    ("Hadi", 75.45, 87.94, 93.13, 2024),
    ("Sri", 75.5, 97.66, 84.81, 2023),
    ("Wawan", 79.13, 72.65, 85.68, 2024),
    ("Yuni", 85.74, 75.88, 82.83, 2022),
    ("Bambang", 82.96, 71.36, 70.76, 2023),
    ("Ayu", 78.74, 79.76, 73.24, 2022),
]

mata_kuliah = ["Algoritma", "Matdis", "AI"]

angkatan_data = {} #mengelompokkan data berdasarkan angkatan
for row in data:
    tahun = row[4]
    nilai = row[1:4]

    if tahun not in angkatan_data:
        angkatan_data[tahun] = [[], [], []]

    for i in range(3): #memasukkan nilai ke masing-masing mata kuliah
        angkatan_data[tahun][i].append(nilai[i])

#melakukan loop untuk setiap angkatan
for tahun in sorted(angkatan_data.keys()):
    plt.figure()

    avg = [np.mean(angkatan_data[tahun][i]) for i in range(3)] #menghitung rata rata nilai tiap mata kuliah
    x = np.arange(len(mata_kuliah))

    plt.bar(x, avg)
    plt.xticks(x, mata_kuliah)

    plt.title(f"Angkatan {tahun}")
    plt.ylabel("Rata-rata Nilai")
    plt.xlabel("Mata Kuliah")
    plt.grid(axis='y')

    plt.show()
```

1. graf kota eropa

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

#mendifinisikan bobot edge
edges = [
    ("Arad", "Zerind", 75), 
    ("Arad", "Sibiu", 140), 
    ("Arad", "Timisoara", 118),
    ("Zerind", "Oradea", 71),
    ("Oradea", "Sibiu", 151),
    ("Sibiu", "Fagaras", 99), 
    ("Sibiu", "Rimnicu Vilcea", 80),
    ("Timisoara", "Lugoj", 111),
    ("Lugoj", "Mehadia", 70),
    ("Mehadia", "Drobeta", 75),
    ("Drobeta", "Craiova", 120),
    ("Craiova", "Pitesti", 138), 
    ("Craiova", "Rimnicu Vilcea", 146),
    ("Rimnicu Vilcea", "Pitesti", 97),
    ("Fagaras", "Bucharest", 211),
    ("Pitesti", "Bucharest", 101),
    ("Bucharest", "Giurgiu", 90),
    ("Bucharest", "Urziceni", 85),
    ("Urziceni", "Hirsova", 98),
    ("Hirsova", "Eforie", 86),
    ("Urziceni", "Vaslui", 142),
    ("Vaslui", "Iasi", 92),
    ("Iasi", "Neamt", 87)
]

G.add_weighted_edges_from(edges)

#menentukan posisi tiap nodenya
pos = {
    "Arad": (1, 5),
    "Zerind": (2, 6),
    "Oradea": (3, 7),
    "Sibiu": (4, 5),
    "Fagaras": (6, 5),
    "Rimnicu Vilcea": (5, 3),
    "Timisoara": (1, 3),
    "Lugoj": (2, 2),
    "Mehadia": (3, 1),
    "Drobeta": (3, 0),
    "Craiova": (5, 0),
    "Pitesti": (6, 2),
    "Bucharest": (8, 2),
    "Giurgiu": (8, 1),
    "Urziceni": (9, 3),
    "Hirsova": (11, 3),
    "Eforie": (12, 2),
    "Vaslui": (10, 5),
    "Iasi": (11, 6),
    "Neamt": (12, 7)
}

#mengatur ukuran figure
plt.figure(figsize=(12,7))
nx.draw(G, pos,
        with_labels=True,
        node_size=1800,
        node_color="lightblue",
        font_size=8,
        font_weight="bold")

labels = nx.get_edge_attributes(G, 'weight')
#menampilkan bobot pada edge
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=7)

plt.title("Graf Kota Eropa (Mirip Gambar Modul)")
plt.axis('off')
plt.show()
```

2.graf pulau jawa

```python
G = nx.Graph()

#mendifinisikan bobot edge
edges = [
    ("Jakarta", "Cirebon", 327),
    ("Jakarta", "Bandung", 270),
    ("Bandung", "Cirebon", 128),
    ("Bandung", "Yogyakarta", 373),
    ("Cirebon", "Semarang", 305),
    ("Cirebon", "Yogyakarta", 210),
    ("Yogyakarta", "Semarang", 109),
    ("Yogyakarta", "Surakarta", 60),
    ("Semarang", "Surakarta", 97),
    ("Semarang", "Surabaya", 369),
    ("Surakarta", "Surabaya", 370),
    ("Surabaya", "Malang", 94)
]

G.add_weighted_edges_from(edges)

#menentukan posisi tiap nodenya
pos = {
    "Jakarta": (0, 3),
    "Bandung": (1, 2),
    "Cirebon": (2.5, 3),
    "Yogyakarta": (4.5, 2),
    "Semarang": (4.5, 3),
    "Surakarta": (5.5, 2),
    "Surabaya": (7.5, 3),
    "Malang": (8.5, 2)
}

#mengatur ukuran figure
plt.figure(figsize=(12,5))
nx.draw(G, pos,
        with_labels=True,
        node_size=1800,
        node_color="lightgreen",
        font_size=9,
        font_weight="bold")

labels = nx.get_edge_attributes(G, 'weight')
#menampilkan bobot pada edge
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=8)

plt.title("Graf Pulau Jawa (Mirip Gambar Modul)")
plt.axis('off')
plt.show()
```
